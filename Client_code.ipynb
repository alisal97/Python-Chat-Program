```python
import time, sys
from socket import *
HEADER_LENGTH = 1000

#---

s = socket()
shost = gethostname ()
ip = gethostbyname (shost)
print (shost, "(", ip, ") \n")
host = input (str ("Enter server address:"))
port = 8081
time.sleep (1)
s.connect ((host, port))
Address = (host,port)
FORMAT = "ASCII"
connect_message=""
disconnect_message ="Disconneted"
my_username = input("Username: ")
first_message = ("POST / HTTP/1.1\n"
         "myline: connect\n"         
         "\n"
         "{}\n"
         "I want to connect").format(my_username)
first_message= first_message.format(my_username)

def connect():
    
    clientSocket = socket(AF_INET,SOCK_STREAM)
    clientSocket.connect(Address)
    return clientSocket



def send(clientSocket, msg):
    """This function send message to the server"""
    mex = msg.encode()
    clientSocket.send(mex)
    
def divide_mex(mex):
    """This functions divides the headers from the content of a message """
    n = 1
    for i in range(len(mex)):
        if mex[i]=="":
            n=i+1
    header = "\n".join(mex[:n])
    mess = "\n".join(mex[n:])
    return header,mess
        
        
def start():
    
    clientSocket = connect()
    send(clientSocket, first_message)
    answer = clientSocket.recv(1024)
    answer = answer.decode("ASCII")
    ans = answer.split(sep="\n")
    header, message = divide_mex(ans)
    print ('From Server:', message)
    
    while clientSocket:
        text = input()
        header = ("POST / HTTP/1.1\n"
        "myline: message to send"
        "\n"
        "\n"
        "{}\n").format(my_username)
        mex = header + text
        
        
        
        send(clientSocket, mex)
        
    send(clientSocket, disconnect_message)
    
start()

    
    
    
```
